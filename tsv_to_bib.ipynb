```python
import pandas as pd
import os
from datetime import datetime
```

```python
talks = pd.read_csv("../Personal_web/markdown_generator/talks.tsv", sep="\t", header=0)
bibfile = "conferences.bib"
talks
```

```python
with open(bibfile, "w") as f:
    for index, row in talks.iterrows():
        key = "".join(row["url_slug"].split(" "))
        bib = f"@unpublished{{{key},\n"
        bib += f"title = {{{row['title']}}},\n"
        bib += f"venue = {{{row['venue']}, {row['location']}}},\n"
        month, year = row["date"].split(" ")
        month = datetime.strptime(month, "%B").strftime("%b")
        bib += f"month = {{{month}}},\n"
        bib += f"year = {{{year}}},\n"
        type = row['type']
        bib += f"keywords = {{{type}}},\n"
        if type != "Attended":
            bib += f"note = {{{row['cont_title']}}},\n"
        bib += f"}}\n"
        f.write(bib)
```
